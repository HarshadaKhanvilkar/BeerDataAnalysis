```python
#Get Review data
beerReviewDf = originalBeerDataDF[["beer_beerId", "review_profileName","review_text"]]

import re
# Removes all special characters and numericals 
def removeSpecialCharacters(text):
    try:
        if pd.isna(text) == False :
            text = re.sub('[^A-Za-z]+', ' ', text)
    except: 
        print("An exception occurred for review ")
        print(text)
    return text

# Cleaning the text in the review column
beerReviewDf["review_text"] = beerReviewDf["review_text"].apply(removeSpecialCharacters)
beerReviewDf.head()
```

```python
# Using NLTK for tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.corpus import wordnet
```

```python
#tags = pos_tag(word_tokenize(text))

cachedStopWords = stopwords.words("english")

def token_postagging_stopwords(text):
    try:
        updatedList = []
        if pd.isna(text) == False :
            tags = pos_tag(word_tokenize(text))
            for word, tag in tags:
                if word.lower() not in cachedStopWords:
                    updatedList.append(tuple([word, tag[0]]))
    except: 
        print("An exception occurred for review ")
        print(text)
    return updatedList
    
def applyPreProcessing(beerReviewDf):
    beerReviewDf["tagged_review"] = beerReviewDf["review_text"].apply(token_postagging_stopwords)
    return beerReviewDf
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

# POS tagger dictionary
pos_dict = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}

def lemmatize(pos_data):
    lemma_rew = " "
    for word, pos in pos_data:
        #print("word = "+word+" pos="+pos)
        if pos is None:
            lemma = word
            lemma_rew = lemma_rew + " " + lemma
        elif pos in pos_dict:
            lemma = wordnet_lemmatizer.lemmatize(word, pos_dict.get(pos))
            lemma_rew = lemma_rew + " " + lemma
    return lemma_rew

```

```python
beerReviewProcessedDf['Lemma'] = beerReviewProcessedDf["tagged_review"].apply(lemmatize)
beerReviewProcessedDf.head()
```

```python
from textblob import TextBlob

# function to calculate subjectivity 
def getSubjectivity(review):
    return TextBlob(review).sentiment.subjectivity

# function to calculate polarity
def getPolarity(review):
    return TextBlob(review).sentiment.polarity

# function to analyze the reviews
def analysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
beerReviewProcessedDf['Polarity'] = beerReviewProcessedDf['Lemma'].apply(getPolarity) 
beerReviewProcessedDf['Analysis'] = beerReviewProcessedDf['Polarity'].apply(analysis)
beerReviewProcessedDf.head()
```
